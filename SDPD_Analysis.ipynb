```python
import numpy as np 
import pandas as pd
```

```python
# a dataset of arouond 100,000 SDPD police stops over the past couple years
data = pd.read_csv("/Users/akalgi/Desktop/SDPD_Data/ripa_stops_datasd.csv")
sdpd_stops = pd.DataFrame(data)
sdpd_stops
```

```python
# drops every unknown address - only about 100 rows, so no significant impact to our data
sdpd_stops = sdpd_stops.dropna(subset="BEATNAME")
sdpd_stops
```

```python
# removes the parenthesis part of each address so we can plot them all later on
def remove_parenthesis(address):
    start_index = address.find("(")
    if start_index == -1:
        return address
    return address[:start_index].strip()
```

```python
# apply the function and keep just the addresses and UID which we will use to merge later
sdpd_addresses = sdpd_stops["BEATNAME"].apply(remove_parenthesis)
sdpd_addresses = pd.DataFrame(sdpd_addresses)
sdpd_addresses["UID"] = sdpd_stops["UID"]
sdpd_addresses
```

```python
# save the dataframe with just addresses of each stop
# this provides us with a method to visually plot all the addresses on a map
sdpd_addresses.to_csv("/Users/akalgi/Desktop/SDPD_Data/sdpd_addresses.csv")
```

```python
# now, lets focus on the overall regions of san diego
def keep_parenthesis(address):
    start_index = address.find("(")
    end_index = address.find(")", start_index)
    if start_index != -1 and end_index != -1:
        return address[start_index:end_index + 1].strip("(").strip(")").strip()
    return ''
```

```python
sdpd_regions = sdpd_stops["BEATNAME"].apply(keep_parenthesis)
sdpd_regions = pd.DataFrame(sdpd_regions)
sdpd_regions["UID"] = sdpd_stops["UID"]
sdpd_regions
```

```python
# this shows us the unique regions in san diego - there are 8 of them
sdpd_regions = sdpd_regions[sdpd_regions.get("BEATNAME") != "Unknown"]
sdpd_regions = sdpd_regions[sdpd_regions.get("BEATNAME") != "Test"]
sdpd_regions.get("BEATNAME").unique()
```

```python
# save the dataframe with regions of each stop
sdpd_regions.to_csv("/Users/akalgi/Desktop/SDPD_Data/sdpd_regions.csv")
```

```python
# now that we have the data for the location of each stop, we want to have the reason of each stop
reason_for_stop_data = pd.read_csv("/Users/akalgi/Desktop/SDPD_Data/ripa_stop_reason_datasd.csv")
sdpd_reason = pd.DataFrame(reason_for_stop_data)
sdpd_reason
```

```python
# note that there are 8 unique reasons for stop
sdpd_reason.get("REASONFORSTOP").unique()
```

```python
# using the reason for stop database, we'll keep the reason for stop as well as the unique identifier
sdpd_reason = sdpd_reason.get(["UID", "REASONFORSTOP"]).drop_duplicates(subset = "UID", keep = "first")
sdpd_reason
```

```python
# do the same process with sdpd addresses and regions, keeping unique UIDs
sdpd_addresses = sdpd_addresses.drop_duplicates(subset = "UID", keep = "first")
sdpd_regions = sdpd_regions.drop_duplicates(subset = "UID", keep = "first")
```

```python
# now merge reason for stop and addresses/regions based on UID
merged_address = sdpd_addresses.merge(sdpd_reason, left_on = "UID", right_on = "UID")
merged_region = sdpd_regions.merge(sdpd_reason, left_on = "UID", right_on = "UID")
```

```python
merged_region
```

```python
# can stop at the above step, but we don't need UID anymore
merged_address = merged_address.drop(columns = "UID")
merged_region = merged_region.drop(columns = "UID")
```
